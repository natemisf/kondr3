```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import pandas as pd
def lotka_volterra(pop, t, alpha, beta, delta, gamma):
   
    x, y = pop
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
   
    return [dxdt, dydt]
alpha = 0.1
beta = 0.02
delta = 0.01
gamma = 0.1

x0 = 40
y0 = 9

initial_pop = [x0, y0]
t = np.linspace(0, 200, 1000)
result = odeint(lotka_volterra, initial_pop, t, args=(alpha, beta, delta, gamma))
df = pd.DataFrame(result, columns=['Жертвы', 'Хищники'])
df['Время'] = t

plt.figure(figsize=(10, 6))
plt.plot(t, df['Жертвы'], label='Жертвы', color='green')
plt.plot(t, df['Хищники'], label='Хищники', color='red')
plt.title('Динамика популяции: Модель Лотки–Вольтерры')
plt.xlabel('Время')
plt.ylabel('Численность')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(df['Жертвы'], df['Хищники'], color='purple')
plt.title('Фазовый портрет (Жертвы vs Хищники)')
plt.xlabel('Жертвы')
plt.ylabel('Хищники')
plt.grid(True)
plt.show()
```
